```python
import os
import matplotlib.pyplot as plt
from email.parser import Parser
from email.utils import parsedate
import time
import numpy as np
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import matplotlib
matplotlib.rc('font', family='serif', size=14)
GMAIL_DIR = "/home/scopatz/gmail/"
```

```python
emails = [e for e in os.listdir(GMAIL_DIR) if e.startswith('0')]
parser = Parser()
timestamps = []
for e in emails:
    with open(os.path.join(GMAIL_DIR, e)) as f:
        m = parser.parse(f, True)
    timestamps.append(time.mktime(parsedate(m['Date'])))
```

```python
ts = np.array(timestamps)
tmin = ts.min()
tmax = ts.max()
n = int((tmax - tmin) / (24*60*60))
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set_ylabel("Emails per day")
ax.hist(ts, bins=n)
ax.set_xlim(tmin, tmax)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos=None: datetime.fromtimestamp(x).strftime('%b %Y')))
fig.autofmt_xdate(rotation=-30, ha="left")
```
